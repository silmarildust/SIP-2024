```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("/content/GRAB APP AND API DATA.xlsx", skiprows=13)
```

```python
m, b = np.polyfit(df['gmaps2_distance (km)'], df['Surge factor'], 1)

x_line = np.linspace(df['gmaps2_distance (km)'].min(), df['gmaps2_distance (km)'].max(), 100)

y_line = m * x_line + b

plt.scatter(df['gmaps2_distance (km)'], df['Surge factor'])
plt.plot(x_line, y_line, color='red')

plt.xlabel('gmaps2_distance (km)')
plt.ylabel('Surge factor')
plt.title('Surge Factor vs. Distance')
plt.show()
```

```python
correlation = df['gmaps2_distance (km)'].corr(df['Surge factor'])
print(correlation)
```
