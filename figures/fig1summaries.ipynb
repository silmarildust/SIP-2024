```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dataplussf.xlsx", skiprows=13)

sns.boxplot(x='Route (API)', y='Surge factor', data=df, flierprops={'marker':'o', 'markerfacecolor':'green', 'markersize':5})
plt.show()
```
